# Get stock data: 

```python
import yfinance as yf 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download('AAPL','2017-01-01','2022-06-01')
data.head()
```

```python
df = data.resample('1q', label='right').last()

df 
#resample the data to quarterly data
```

```python
df.describe().round(2)
```

# Calculate returns: 

```python
adj_close = df['Adj Close']
#adj_close
df = adj_close
df
```

```python
df.to_excel('AAPL_ADJ_Close.xlsx')
```

```python
abs_dif = adj_close.diff()
df= abs_dif
df
```

```python
df.to_excel('AAPL_Absolute_change.xlsx')
```

```python
pct_change = abs_dif.pct_change().round(4)

df= pct_change
df
#percentage returns 
```

```python
df.to_excel('AAPL_Perc_change.xlsx')
```

```python
rets = np.log(adj_close / adj_close.shift(1))
#rets.round(4)
df = rets.round(4)

df
#log returns 
```

```python
df.to_excel('AAPL_Log_Ret.xlsx')
```
