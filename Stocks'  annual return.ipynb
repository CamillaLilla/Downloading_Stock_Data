```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('PORTFOLIO_ADJ_CLOSE.xlsx') 

#this spreadsheet contains adj_close for each stock that need to be analysed.

df
```

```python
Ret= df.pct_change().apply(lambda x: np.log(1+x))
```

```python
Ann_Ret = Ret.mean()*52 
```

```python
Ann_Risk = Ret.std()*(52)**(0.5) 
```

```python
Ret_Risk = (Ann_Ret/Ann_Risk) 
```

```python
df = Ret_Risk 
```

```python
df.to_excel('annual_return.xlsx')
```
